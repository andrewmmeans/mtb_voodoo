```python
def ride_log_summary(soup):
    ride_log_sum_dict = {
        'rides_past_week':0,
        'rides_six_months':0,
        'rides_total':0
    }
    try:
        ride_log_sum_dict['rides_past_week'] = int(soup.select('.block > div:nth-child(1) > div:nth-child(1) > ul:nth-child(2) > li:nth-child(1) > b:nth-child(1)')[0].text)
    except:
        pass
    try:
        ride_log_sum_dict['rides_six_months'] = int(soup.select('.block > div:nth-child(2) > div:nth-child(1) > ul:nth-child(2) > li:nth-child(1) > b:nth-child(1)')[0].text)
    except:
        pass
    try:
        ride_log_sum_dict['rides_total'] = int(soup.select('.block > div:nth-child(3) > div:nth-child(1) > ul:nth-child(2) > li:nth-child(1) > b:nth-child(1)')[0].text)
    except:
        pass
    return ride_log_sum_dict



```

```python
ride_log_summary(soup)
```

```python
def extract_trail_ids_from_tag(tag):
    url = tag.attrs.get('href')
    return url
```

```python
table = soup.find('section').find('table',id='').find('tbody').find_all('tr')

ride_date = []
rider_url = []
rider_name = []
trails_ridden = []
ride_difficulty_string = []


for row in table:
    row_struct = row.find_all('td')
    ride_date.append(row_struct[1].find('span').text)
    ride_date = [x + ', 2020' if ',' not in x else x for x in ride_date]
    rider_url.append(row_struct[2].find('a').attrs['href'])
    rider_name.append(row_struct[2].find('a').text)
    trail_urls = []
    for tag in row_struct[4]:
        if type(tag) is Tag:
            trail_urls.append(extract_trail_ids_from_tag(tag))
    trails_ridden.append(trail_urls)
    ride_difficulty_string.append(row_struct[5].span.attrs['title'])

#trails_ridden = [x.strip().split(',\n') for x in trails_ridden]
```
